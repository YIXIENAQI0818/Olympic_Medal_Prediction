```python
import pandas as pd

patch_df = pd.read_csv('..//Data//patch02.csv')
medal = pd.read_csv('..//Data//summerOly_medal_counts.csv')


patch_df.drop(columns=['medal_gold', 'medal_silver', 'medal_bronze'], inplace=True)
patch_df['medal_gold'] = 0
patch_df['medal_silver'] = 0
patch_df['medal_bronze'] = 0


updated_rows = []

for _, row_patch in patch_df.iterrows():
    # 转换 row_patch 为字典以便修改
    updated_row = row_patch.to_dict()
    
    for _, row_medal in medal.iterrows():
        if row_medal['NOC'] == row_patch['country'] and row_medal['Year'] == row_patch['year']:
            updated_row['medal_gold'] = row_medal['Gold']
            updated_row['medal_silver'] = row_medal['Silver']
            updated_row['medal_bronze'] = row_medal['Bronze']
    
    # 将更新后的行数据存储
    updated_rows.append(updated_row)

# 将列表转换为新的 DataFrame
new_patch_df = pd.DataFrame(updated_rows)

print(new_patch_df)
```

```python
new_patch_df.to_csv('..//output//patch_medal02.csv', index=False)
```
